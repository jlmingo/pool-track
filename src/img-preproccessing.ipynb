```python
def img_to_pool(preprocessed_image):
    img = cv2.imread(preprocessed_image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    l_blue = np.array([75,50,50])
    d_blue = np.array([150,255,255])
    mask = cv2.inRange(hsv_img, l_blue, d_blue)
    result = cv2.bitwise_and(img, img, mask=mask)
    return result

def is_pool(processed_image, path):
    grid = np.empty((4,4))
    for i in range(4):
        for j in range(4):
            grid[i,j] = np.mean(processed_image[64*i:64*i+64, 64*j:64*j+64])
    max_index = np.where(grid == np.amax(grid))
    max_index = np.resize(np.stack(max_index), (1,2))
    max_value = np.amax(grid)
    return (max_index, max_value, path) if max_value>5 else None

def poolDetect(path):
    files = glob.glob(path+"*.png")
    detected_pools = []
    for f in files:
        result = img_to_pool(f)
        if is_pool(result, f):
            detected_pools.append(is_pool(result, f))
    return detected_pools
```
